```python
df_total_por_ano = df_plot.groupby('Ano')['Quantidade de Nascimentos'].sum().reset_index()
df_total_por_ano['Quantidade de Nascimentos (Milhares)'] = df_total_por_ano['Quantidade de Nascimentos'] / 1000

plt.figure(figsize=(10, 6))
sns.barplot(
    x='Ano',
    y='Quantidade de Nascimentos (Milhares)',
    data=df_total_por_ano,
    palette='viridis'
)

plt.title('Total de Nascimentos ao Longo dos Anos no Brasil')
plt.xlabel('Ano')
plt.ylabel('Total de Nascimentos (Milhares)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
